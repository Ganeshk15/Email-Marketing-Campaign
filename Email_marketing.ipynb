```python
numerical_columns=data.select_dtypes(include=['number']).columns
numerical_columns
```

```python
# outlier detection
def outlier_detection(df,colname): #creating function to finding outiers in each columns
    #calculate the Q1,Q3 and IQR
    q1=df[colname].quantile(0.25)
    q3=df[colname].quantile(0.75)
    iqr=q3-q1

    upper_extreme=q3+(1.5 * iqr)
    lower_extreme=q1-(1.5 * iqr)

    # Identify outliers
    outliers = df[(df[colname] < lower_extreme) | (df[colname] > upper_extreme)]


    return lower_extreme,upper_extreme,q1,q3,outliers
    
```

```python
outlier_detection(data,'Purchase_History') #lower_extrem,upper_extreme,q1,q3,outliers of  'Purchase_History'
```

```python
#printing number of outliers in each columns
for col in numerical_columns:
    #calling the outlier detection function
    _,_,_,_, outliers=outlier_detection(data,col)

    print(f"Number of outliers in '{col}': {len(outliers)}")
```

```python
#handling outliers by capping them
def handle_outliers(df,column):
    lower_extreme, upper_extreme,_,_,_=outlier_detection(data,column)
    df[column]=df[column].clip(lower=lower_extreme,upper=upper_extreme)
    #Any value less than the lower_extreme will be replaced by lower_extreme.
    #Any value greater than the upper_extreme will be replaced by upper_extreme.
    return df
    
for column in numerical_columns:
    data=handle_outliers(data,column)
    print(f"outliers replaced in the column :{column}")
    
```

# 2. Model Bulding

```python
# Split the Data into Training and Testing Sets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#split the data in to features(x) and target(y=opened_previous_emails)
x=data.drop(columns=['Opened_Previous_Emails']) # all cols except target variable
y=data['Opened_Previous_Emails'] #target variable

#standardization (only numerical coloumn not the binary ones)
binary_columns=['Opened_Previous_Emails','Clicked_Previous_Emails','Device_Type']

#seperate numerical cols to standardize
standardizing_columns=[col for col in x.columns if col not in binary_columns]
print(standardizing_columns)
```

The train-test splits the dataset into features (x) and target (y), and identifies which columns are numerical (excluding binary columns) to standardize. It then prints the list of columns that will be standardized, excluding binary ones.

```python
# Apply StandardScaler to numerical columns only (exclude binary columns)
scaler = StandardScaler()
x[standardizing_columns]=scaler.fit_transform(x[standardizing_columns])
x #standized features table
```

```python
#split the data into training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape
```

## 2.1 Neural Networks

```python
!pip install tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Build the Neural Network model using Keras
model_1 = Sequential()
```

```python
# Add an input layer with the number of features in X
model_1.add(Dense(64, input_dim=x_train.shape[1], activation='relu'))
```

```python
# Add a hidden layer
model_1.add(Dense(32, activation='relu'))
```

```python
# Add the output layer with a sigmoid activation function for binary classification
model_1.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model with binary crossentropy loss and an Adam optimizer
model_1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
history = model_1.fit(x_train, y_train, epochs=25, batch_size=32, validation_data=(x_test, y_test))
```

Training a machine learning model (model_1) using the training data (x_train, y_train) for 25 epochs, with a batch size of 32, while also validating the model performance on the test data (x_test, y_test) after each epoch. The history object stores the training and validation loss/accuracy for each epoch.

```python
# Evaluate the model on the test data
y_pred1 = model_1.predict(x_test)
# Convert probabilities to binary predictions
y_pred1 = (y_pred1 > 0.5).astype(int)
y_pred1
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Evaluate model performance using accuracy, confusion matrix, and classification report
print("Accuracy Score:", accuracy_score(y_test, y_pred1))
print("Classification Report:\n", classification_report(y_test, y_pred1))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred1))
```

```python
# Visualizing confusion matrix
plt.figure(figsize=(12,6))
sns.heatmap(confusion_matrix(y_test, y_pred1), annot=True, fmt='d')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

The confusion matrix for the test data (y_test) and the predicted labels (y_pred1) using confusion_matrix().

## 2.2 RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

#initialize model
model_2 = RandomForestClassifier(n_estimators=100, random_state=42)

#Train the model on training data
model_2.fit(x_train,y_train)

#predict on test data
y_pred2=model_2.predict(x_test) 
```

```python
# Visualizing confusion matrix
plt.figure(figsize=(12,6))
sns.heatmap(confusion_matrix(y_test, y_pred2), annot=True, fmt='d')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

The confusion matrix for the test data (y_test) and the predicted labels (y_pred2) using confusion_matrix().

```python
# Evaluating the model
accuracy_rf = accuracy_score(y_test, y_pred2)
precision_rf = precision_score(y_test, y_pred2, average='weighted')
recall_rf = recall_score(y_test, y_pred2, average='weighted')
f1_rf = f1_score(y_test, y_pred2, average='weighted')
# print the values
print(f"Accuracy: {accuracy_rf}")
print(f"Precision: {precision_rf}")
print(f"Recall: {recall_rf}")
print(f"F1 Score: {f1_rf}")
```
